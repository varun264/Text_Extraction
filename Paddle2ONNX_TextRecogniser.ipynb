Convert Paddle static graph model to ONNX model format using Paddle2ONNX:

```python
#convering PaddleOCR detection model into ONNX detection model
#db_det
!paddle2onnx --model_dir /content/PaddleOCR/inference/det_db \
--model_filename inference.pdmodel \
--params_filename inference.pdiparams \
--save_file ./inference/det_onnx/model.onnx \
--opset_version 10 \
--input_shape_dict="{'x':[-1,3,-1,-1]}" \
--enable_onnx_checker True


```

```python
#fce_det
!paddle2onnx --model_dir /content/PaddleOCR/PaddleOCR/inference/det_fce \
--model_filename inference.pdmodel \
--params_filename inference.pdiparams \
--save_file ./inference/det_fce_onnx/model.onnx \
--opset_version 10 \
--input_shape_dict="{'x':[-1,3,-1,-1]}" \
--enable_onnx_checker True
```

This error has come because fce is not yet supported on ONNX format

```python
#Converting the text recogniser PaddleOCR model into ONNX text recogniser model
%cd PaddleOCR
!paddle2onnx --model_dir ./inference/CRNN_mv3 \
--model_filename inference.pdmodel \
--params_filename inference.pdiparams \
--save_file ./inference/rec_onnx/model.onnx \
--opset_version 10 \
--input_shape_dict="{'x':[-1,3,-1,-1]}" \
--enable_onnx_checker True
```

```python
#this is for a PaddleOCR classifier to ONNX classifier
paddle2onnx --model_dir ./inference/ch_ppocr_mobile_v2.0_cls_infer \
--model_filename ch_ppocr_mobile_v2.0_cls_infer/inference.pdmodel \
--params_filename ch_ppocr_mobile_v2.0_cls_infer/inference.pdiparams \
--save_file ./inferencecls_onnx/model.onnx \
--opset_version 10 \
--input_shape_dict="{'x':[-1,3,-1,-1]}" \
--enable_onnx_checker True
```

Inference prediction

```python
#ONNX runtime
%cd PaddleOCR
!python3.7 tools/infer/predict_system.py --use_gpu=False --use_onnx=True \
--det_model_dir=./inference/det_onnx/model.onnx  \
--rec_model_dir=./inference/rec_onnx/model.onnx  \
--image_dir=/content/test.jpeg \
--rec_image_shape="3, 32, 320" --use_angle_cls=false --rec_char_dict_path="./ppocr/utils/en_dict.txt"
```
